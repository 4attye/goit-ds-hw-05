6.Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

```python
columns_with_nan = df.columns[df.isnull().any()]
columns_to_drop = [col for col in columns_with_nan if col != "Мова програмування"]
df_cleaned = df.drop(columns=columns_to_drop)
df_cleaned
```

10.Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

```python
python_data = df_dropna[df_dropna['Мова програмування'] == 'Python']
python_data
```

12.Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
grouped = python_data.groupby('Посада')
```

13.Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
salary = grouped['Зарплата на місяць'].agg(['min', 'max'])
salary
```

14.Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(row):
    return row.mean()

salary['avg'] = salary.apply(fill_avg_salary, axis=1)
salary
```

15.Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
salary['avg'].describe()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Місто'], df['Зарплата на місяць'], color='purple')
plt.title('Зарплата в залежності від міста')
plt.xlabel('Досвід (років)')
plt.ylabel('Зарплата в місяць')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
programming_languages = df['Мова програмування'].value_counts().head(10)

plt.figure(figsize=(8, 8))
plt.pie(programming_languages, labels=programming_languages.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Частка мов програмування серед розробників')
plt.tight_layout()
plt.show()
```

```python
avg_salary_by_position = df.groupby('Посада')['Зарплата на місяць'].mean()

plt.figure(figsize=(15, 8))
avg_salary_by_position.plot(kind='bar', color='coral')
plt.title('Середня зарплата за посадами')
plt.xlabel('Посада')
plt.ylabel('Середня зарплата')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
